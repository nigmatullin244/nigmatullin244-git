### Как Git отслеживает статус файлов

Чтобы файлы были видны Git и сохранились в репозитории их необходимо отслеживать (git add)

Отслеживаемые файлы существуют в одном из трех состояний:
    1. Сохраненное - файл в рабочей копии идентичен последнему коммиту
    2. Измененное - файл в рабочей копии отличается от  последнего коммита
    3. Индексированное - файл изменен и помечен для включения в следующий коммит

Как файлы переходят из состояний:
    1. Неотслеживаемый -> git add -> Отслеживаемый
    2. Сохраненный -> изменение файла -> Измененный 
    3. Измененный -> git add -> Индексированный
    4. Измененный -> git commit -> Сохраненный

Команда git status - просмотр статуса текущего репозитория и его файлов


### Проблемы индексирования

Файл может быть изменен после того, как он был проиндексирован, в результате чего файл существует как в измененном, так и в индексированном состоянии.
Индексирование описывает описание изменений файлов, и одни части измененного файла могут быть индексированны, а другие нет.

Как упростить работу с индексацией:
    1. Использовать команду git commit -am command
        - Файлы переходят из измененного состояния сразу же в чистое.
    2. Сохранение файлов сразу после их их добавления, переименования или удаления из репозитория.
    3. Использование графических средств Git

### Восстановление старых изменений

1. Отмена несохраненных локальных изменений
    - git restore <имя_файла> -  отмена изменений, внесенных в файл (который еще не был проиндексирован или сохранен)
    - нельзя отменить git restore, чтобы вернуть последние изменения
    - git checkout . - отмена всех изменений во всех файлах рабочей копии
2. Деиндексирование проиндексированного файла
    - git restore --staged <имя_файла> - исключение из индексированного состояния
3. Отмена последних коммитов
    - git revert -n HEAD~<число>..HEAD. - отмена последних коммитов
4. Возврат к конкретному коммиту для отдельного файла
    - git show <хеш>: <имя_файла> - показывает файл на момент последнего конкретного коммита
    - git checkout <хеш> -- <имя_файла> - возвращает содержимое файла к этой версии.
5. Перезапись истории коммитов
    - git filter-branchor - проходится по коммита и изменяет их
    - git filter-repo - тоже самое, но более безопасно 
    - BFG Repo-Cleaner — утилита для очистки истории Git-репозиториев от нежелательных данных, таких как пароли, ключи доступа, API-токены и слишком большие файлы. Она выступает как простой и быстрый аналог команды git filter-branch.

### Анализ сделанных изменений

git diff - показывает изменения

Сама команда не просто выводит на экран разницу между файлами, но и запускает пейджер — специальную программу, которая позволяет перемещаться по выводу и искать внутри него нужные данные. Для этого используются клавиши:

f — перемещение вниз
b или u — перемещение вверх
q — или выход из режима просмотра


По умолчанию команда git diff показывает изменения только для тех модифицированных файлов, которые еще не были добавлены в индекс. Подразумевается, что добавленные в индекс файлы смотреть не нужно, ведь мы их уже подготовили к коммиту. Но лучше еще раз увидеть эти изменения и все проверить. Для этого нужно запустить команду git diff с флагом --staged

### Отмена изменений в рабочей директории

1. Неотслеживаемые файлы
    - git clean -fd - удаление неотслеживаемых файлов (-f – force, -d – directory)
2. Измененные файлы в рабочей директории
    - git restore файл - восстанавливает рабочий каталог (рабочую копию файла) из индекса, возвращая файл к последнему состоянию, зафиксированному в индексе. То есть все незакоммиченные изменения в рабочем файле будут отменены, и файл станет таким, каким он был подготовлен к коммиту.
3. Изменения, подготовленные к коммиту
    - git restore --staged <file> отменяет индексацию изменений (unstage), то есть убирает файл из области подготовленных для коммита (staging area), но оставляет изменения в рабочем каталоге без изменений. Проще говоря, изменения остаются в файле, но теперь они не будут включены в следующий коммит. 

### Отмена коммитов
1. Отмена изменений - создание еще одного коммита, который выполняет изменения, противоположные тому коммиту, который отменяется
    - git revert <хэш коммита на который хотите вернуться>
    - Команда revert может отменять не только последний коммит, но и любой другой коммит из истории проекта
2. Нужно удалить только что сделанный по ошибке коммит
    - git reset
    - git reset --hard HEAD~ - hard полное удаление
    - HEAD~ - один коммит от последнего коммита
    - HEAD~2 - 2 последних коммита
    - Без --hard команда git reset по умолчанию работает в режиме --mixed, при котором изменения остаются в рабочей директории, но исключаются из индекса (unstage). Затем их можно исправить или отменить и выполнить новый коммит
    - Команда git reset --soft отменяет коммит, но оставляет изменения в индексе (staging area). Следующий коммит включит в себя те же изменения, если их не модифицировать.

### Изменение последнего коммита
Крайне часто разработчики делают коммит и сразу понимают, что забыли добавить часть файлов через git add. Оставшуюся часть изменений можно дослать следующим коммитом.

Есть еще один способ. Если изменения еще не были отправлены во внешнюю систему, можно добавить изменения в текущий коммит. Для этого во время коммита добавляется флаг --amend

В реальности --amend не добавляет изменения в существующий коммит. Этот флаг приводит к откату коммита через git reset и выполнению нового коммита с новыми данными. Поэтому мы и видим ровно один коммит, хотя команда git commit выполнялась два раза (первый раз — когда сделали ошибочный коммит).
git commit --amend можно добавить опцию --no-edit - чтобы не изменялось описание коммита
